```python
import pandas as pd

# Load the uncalibrated and calibrated data
uncalibrated_file_path = '/content/RData/RoadAnomalies/3.Bump/AccelerometerUncalibrated.csv'  # Replace with your actual file path
calibrated_file_path = '/content/RData/RoadAnomalies/3.Bump/Accelerometer.csv'      # Replace with your actual file path

uncalibrated_data = pd.read_csv(uncalibrated_file_path)
calibrated_data = pd.read_csv(calibrated_file_path)

# Assuming the sensor was stationary at the beginning of the recording
# Extracting the first few seconds of stationary data for offset calculation
# Adjust the condition based on your data
stationary_duration = 5  # Duration in seconds for stationary data
stationary_data = uncalibrated_data[uncalibrated_data['seconds_elapsed'] <= stationary_duration]

# Calculate offsets for each axis
offset_x = stationary_data['x'].mean()
offset_y = stationary_data['y'].mean()
offset_z = stationary_data['z'].mean()

# Print the calculated offsets
print("Calculated Offsets from Stationary Data:")
print(f"Offset X: {offset_x}")
print(f"Offset Y: {offset_y}")
print(f"Offset Z: {offset_z}")

# Optionally, you can compare these offsets with the calibrated values
# For instance, get the mean of the calibrated data during the same time frame
calibrated_stationary_data = calibrated_data[calibrated_data['seconds_elapsed'] <= stationary_duration]

calibrated_mean_x = calibrated_stationary_data['x'].mean()
calibrated_mean_y = calibrated_stationary_data['y'].mean()
calibrated_mean_z = calibrated_stationary_data['z'].mean()

print("\nMean Calibrated Values from Stationary Data:")
print(f"Calibrated Mean X: {calibrated_mean_x}")
print(f"Calibrated Mean Y: {calibrated_mean_y}")
print(f"Calibrated Mean Z: {calibrated_mean_z}")

# You can also generate a summary DataFrame for better presentation
summary_data = pd.DataFrame({
    'Axis': ['X', 'Y', 'Z'],
    'Offset (Uncalibrated)': [offset_x, offset_y, offset_z],
    'Mean Calibrated Value': [calibrated_mean_x, calibrated_mean_y, calibrated_mean_z]
})

print("\nSummary of Offsets and Calibrated Values:")
print(summary_data)
```
